```python
# Data Visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
%matplotlib inline

# Pre-processing
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder

# Model
import keras
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from tensorflow.keras.optimizers import Adam         # Adaptive Moment Estimation
from keras.layers import LeakyReLU

# Metrics
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
data = pd.read_csv('drive/My Drive/Colab Notebooks/csv_dataset/Churn_Modelling.csv')
data.head()
```

```python
X = data.iloc[:, 3:-1].values
y = data.iloc[:, -1].values
```

Convert the gender where instead of "Female" and "Male", we make it 0 and 1 using label encoder function.

```python
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])
print(X)
```

```python
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [1])], remainder = 'passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

Split the training and testing datasets. (80/20) and transform it

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 20)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## **Building the Model**

```python
# Initialize the ANN model
ann = tf.keras.models.Sequential()

# Input layer & first hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='LeakyReLU'))
# Second hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='LeakyReLU'))
ann.add(tf.keras.layers.Dense(units=6, activation='LeakyReLU'))
ann.add(tf.keras.layers.Dense(units=6, activation='LeakyReLU'))
# Output layer
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Compile the model
ann.compile(optimizer= 'adagrad', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# train the model
ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

Testing datasets prediction result and output the accuracy

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
accuracy_score(y_test, y_pred)
```
