**Step 2.** Import necessary libraries.

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

**Step 3.** Load and display the image.

```python
image = cv2.imread('/content/drive/My Drive/dataset/tiger.jpg')

# convert to RBG
image2 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
 
# for plotting the two images
images = [image, image2]
labels = ['Loaded Image', 'Converted to RGB']
fig = figure(figsize=(20, 10))

for i in range(0, 1*2):
  fig.add_subplot(1, 2, i+1)
  plt.title(labels[i], style="oblique")
  plt.imshow(images[i])
plt.show()
```

**Step 4.** Convert the image to grayscale.

```python
# converting the image into grayscale
gray = cv2.cvtColor(image2, cv2.COLOR_RGB2GRAY)

images = [image2, gray]
labels = ['Image', 'Converted to Grayscale']
fig = figure(figsize=(20, 10))

for i in range(0, 1*2):
  fig.add_subplot(1, 2, i+1)
  plt.title(labels[i], style="oblique")
  plt.imshow(images[i], cmap='gray')
plt.show()
```

**Step 5.** Blur the image using Gaussian Blur for better edge detection.


```python
# Blur the image for better edge detection
img_blur = cv2.GaussianBlur(gray, (3,3), sigmaX=0, sigmaY=0)

images = [gray, img_blur]
labels = ['Grayscale', 'Grayscale + Gaussian Blur']
fig = figure(figsize=(20, 10))

for i in range(0, 1*2):
  fig.add_subplot(1, 2, i+1)
  plt.title(labels[i], style="oblique")
  plt.imshow(images[i], cmap='gray')
plt.show()
                                                 
```

**Step 6.** Detect edges of the image using Sobel. Create a 3x3 array for edge detection

The sobel filter uses two 3 x 3 kernels. One for changes in the
_horizontal_ direction, and one for changes in the _vertical_ direction.
The two kernels are convolved with the original image to calculate the
approximations of the derivatives.

```python
# 3x3 kernel for vertical direction -> y-axis
sobel_y = np.array([[ -1, -2, -1], 
                   [ 0, 0, 0], 
                   [ 1, 2, 1]])

# 3x3 kernel for horizontal direction -> x-axis
sobel_x = np.array([[-1, 0, 1],
                    [-2,0,2],
                    [-1, 0, 1]])
```

**Step 7.** Filter the image using filter2D. It accepts 3 parameters: grayscale image, bit-depth, and kernel

```python
filtered_image = cv2.filter2D(gray, -1, img_blur)

# Y-axis
filtered_image_y = cv2.filter2D(gray, -1, sobel_y)

# X-axis
filtered_image_x = cv2.filter2D(gray, -1, sobel_x)
```

**Step 8.** Display the output of applying Sobel filter in the x-axis.

```python
figure(figsize=(20, 10))
plt.imshow(filtered_image_x, cmap='gray')
```

**Step 9.** Display the output of applying Sobel filter in the y-axis.

```python
figure(figsize=(20, 10))
plt.imshow(filtered_image_y, cmap='gray')
```

**Step 10.** Sum both of the images and display the combined output. 

```python
final_image = cv2.add(filtered_image_x, filtered_image_y)
figure(figsize=(20, 10))
plt.imshow(final_image, cmap='gray')
```
